```python
# importing the libraries needed
import os     
import random 
```

```python
def thanos_project(folder_path, num_files):
     # 1- create folder if not exists 
    if not os.path.exists(folder_path):
        os.mkdir(folder_path)

    # 2- create number of files inside it
    for i in range(num_files):
        file_path = os.path.join(folder_path, f"file_{i}.txt")
        open(file_path, 'w').close()

    # 3- check the len of files in that folder
    files = os.listdir(folder_path)
    print("Number of files before delete:", len(files))

    # 4- delete random half of this folder
    files_to_delete = random.sample(files, len(files)//2)
    for file in files_to_delete:
        os.remove(os.path.join(folder_path, file))

    # 5- check the result
    final_files = os.listdir(folder_path)
    print("Number of files after delete:", len(final_files))
```

```python
 # tryingg
thanos_project("thanos_folder", 10)
```
