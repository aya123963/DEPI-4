```python
# Decoding message, extracting the core part

import re

def extract_core(s):
    return re.findall(r'[A-Za-z]+', s)

x = "###!!@mocleW EPGTQ!!!6789"
core = extract_core(x)
print(core)
```

```python
# reversing the word 

def decode_first_word(message):
   
    words = message.split()
    
   
    return words[0][::-1]


encoded = "emocleW EPGTQ"
decoded = decode_first_word(encoded)
print(decoded)
```

```python
# Replacing vowels

def vowel_shift(message):
   
    words = message.split()
    
    
     first_word = words[0][::-1]
    
    
    vowels = "AEIOUaeiou"
    second_word = ""
    for char in words[1]:
        if char in vowels:
           
            second_word += "A"
        else:
            second_word += char
    
   
    return first_word + " " + second_word


encoded = "emocleW EPGTQ"
decoded = vowel_shift(encoded)
print(decoded)
```

```python
# Decode message 

def decode_message(message):
    
    words = message.split()
    first_word = words[0][::-1]
    vowels = "AEIOUaeiou"
    second_word = ""
    for char in words[1]:
        if char in vowels:
            second_word += char
        else:
            second_word += char
    return first_word + " " + second_word
encoded = "emocleW EPGTQ"
print(decode_message(encoded))
```
