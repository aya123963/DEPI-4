```python
# Extracting the core part

message = "yalpstcejorp EPUVT"
words =message.split()
decoded = [word[::-1] for word in words]
result = " ".join(decoded)

print(result)
```

```python
#  Reversing the first word

x = "yalpstcejorp EPUVT"
words = x.split()
words[0] = words[0][::-1]

result = " ".join(words)
print(result)
```

```python
# Replacing the shifted vowels

y= "projectplay EPUVT"
words = y.split()
replace_map = {
    "E": "A",
    "U": "O"
}
new_word = ""
for ch in words[1]:
    new_word += replace_map.get(ch, ch)

words[1] = new_word

result = " ".join(words)
print(result)
```

```python
# Decoding the final message

v= "projectplay APOVT"
words =v.split()
words[1] = "APTOV"

final_message = " ".join(words)
print(final_message)
```
