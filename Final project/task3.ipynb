```python
from abc import ABC, abstractmethod

# creating the abstract class
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        """Abstract method: Each animal class  must implement this function """
        pass

    def describe(self):
        """Concrete method: general description of the current animal"""
        print(f"I am a {self.__class__.__name__} and I am an animal")

# creating the subClasses derived from the Animal abstract class by passing the abstract class name as an attribute to each subclass
class Dog(Animal):
    def make_sound(self):
        return "woof!"

class Cat(Animal):
    def make_sound(self):
        return "meow!"

class Cow(Animal):
    def make_sound(self):
        return "moo! "

class Bird(Animal):
    def make_sound(self):
        return "seeww!"

#  the final step is to test the classes
if __name__ == "__main__":
    animals = [Dog(), Cat(), Cow(), Bird()]
    
    for animal in animals:
        animal.describe()
        print("Sound:", animal.make_sound())
        print("-" * 30)
```
