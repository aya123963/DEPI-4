```python

#  creating the function to read text file
def read_txt_file(file_path):
    """
    reads the contents of a text file and returns it as a string

    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return f"error,  file '{file_path}' was not found"
    except IOError:
        return "error occurred while reading the file"

# creating a  class to extract usernames
class UserExtractor:
    def __init__(self, file_path):
        """
        intializes the class with file_path and an empty usernames dictionary
        """
        self.file_path = file_path
        self.usernames = {}  # I created a dictionary to store usernames 

    def extract_usernames(self):
        """
        reads the file using read_txt_file and extracts usernames into the dictionary
       
        """
        content = read_txt_file(self.file_path)

        # checking if an error occurred during file reading
        if isinstance(content, str) and content.startswith("Error:"):
            return content  

        # processing each line
        lines = content.splitlines()
        for line in lines:
            # ignoring empty lines
            if not line.strip():
                continue

            # spliting by ':' and extract username and password
            if ':' in line:
                username, password = line.split(':', 1)
                username = username.strip()
                password = password.strip()
                #making sure that usernames are unique
                if username not in self.usernames:
                    self.usernames[username] = password
            else:
                print(f"Warning: Line skipped due to incorrect format -> '{line}'")

        return self.usernames


if __name__ == "__main__":
    file_path = "users.txt"  

    extractor = UserExtractor(file_path)
    result = extractor.extract_usernames()

    print("Extracted usernames and passwords:")
    print(result)
```
