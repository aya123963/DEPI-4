```python
class BankAccount:    #  creating class named BankAccount to gather all of the methods : desposit,withdraw,checkbalance 
    def __init__(self, initial_balance):  # basic constructor to initialize the  class
        if initial_balance < 0:   # majing sure that the client doesn't put any negative initial value to the account 
            raise ValueError("Initial balance cannot be negative")
        self.balance = initial_balance

    def deposit(self, amount):   # this function is responsible for adding the amount that the client deposites to the current account balance
        if amount <= 0:
            print("Deposit amount must be positive")
        else:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")

    def withdraw(self, amount):    # this function is responsible for withdrawing amount from the current balance 
        if amount <= 0:
            print("Withdrawal amount must be positive")
        elif amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")

    def check_balance(self):         # this function is responsible for checking the current balance
        print(f"Current balance: {self.balance}")
```

```python
account = BankAccount(20)     # initializing the class with initial value
```

```python
account.deposit(50)
```

```python
account.withdraw(30)
```

```python
account.check_balance()
```

```python
account.deposit(-10)
account.withdraw(500)
```
