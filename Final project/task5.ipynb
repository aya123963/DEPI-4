```python
def unique_words_count(list):
    """
    this function takes a list of words and returns  dictionary wth 
    the number of times  of each unique word
    """
    unique_words = set(list)
    counts = {word: 0 for word in unique_words}
    
    for word in list:
        counts[word] += 1
        
    return counts
```

```python
words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]

result = unique_words_count(words)
print(result)
```
