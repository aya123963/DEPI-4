```python
class TextFileReader:
    def __init__(self, file_path):
        """
        Initializes the TextFileReader with the path to the file.
        """
        self.file_path = file_path
        self.content = ""  # To store the content after reading

    def read_file(self):
        """
        Reads the content of the file and stores it in self.content.
        """
        try:
            with open(self.file_path, 'r', encoding='utf-8') as file:
                self.content = file.read()
            print("File read successfully!")
        except FileNotFoundError:
            print(f"Error: The file '{self.file_path}' was not found.")
        except Exception as e:
            print(f"An error occurred: {e}")

    def count_lines(self):
        """
        Returns the number of lines in the file.
        """
        return len(self.content.splitlines())

    def count_words(self):
        """
        Returns the total number of words in the file.
        """
        return len(self.content.split())

    def count_characters(self):
        """
        Returns the total number of characters in the file.
        """
        return len(self.content)

    def display_content(self):
        """
        Prints the content of the file.
        """
        print(self.content)


# ----------------------------
# Example usage
if __name__ == "__main__":
    # Replace 'example.txt' with the path to your text file
    reader = TextFileReader("example.txt")
    
    reader.read_file()
    
    print("Number of lines:", reader.count_lines())
    print("Number of words:", reader.count_words())
    print("Number of characters:", reader.count_characters())
    
    print("\nFile content:")
    reader.display_content()
```
