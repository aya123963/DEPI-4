```python
# Extracting the core part of the encoded message
def core_message(message):
    core = ""
    for char in message:
        if char.isalpha() or char == " ":
            core += char
    return core.strip()
x = "&&&**$gnirtS PLIO!!@1234"
print(core_message(x))
```

```python
# Reverse the first word
def reverse_first_word(message):
    words = message.split()
    return words[0][::-1]
core_message = "gnirtS PLIO"
print(reverse_first_word(core_message))
```

```python
# Replacing the shifted vowels

def replace_shifted_vowels(word):
    replacements = {
        "I": "E",
        "O": "U"
    }

    result = ""
    for char in word:
        if char in replacements:
            result += replacements[char]
        else:
            result += char

    return result
second_word = "PLIO"
print(replace_shifted_vowels(second_word))
```

```python
#  Final decoded message 
def final_decoded_message(message):
    words = message.split()
    first_word = words[0][::-1]
    replacements = {"I": "E", "O": "U"}
    second_word = ""
    for char in words[1]:
        second_word += replacements.get(char, char)
    return first_word + " " + second_word
core_message = "gnirtS PLIO"
print(final_decoded_message(core_message))
```
