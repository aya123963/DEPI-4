```python
class SetItemRemover:
    """
    SetItemRemover class handles the operation of removing an item from a set.

    Attributes:
    my_set (set): The set from which an item will be removed.
    item (any): The item to be removed from the set.
    """

    def __init__(self, my_set, item):
        """
        Initializes the SetItemRemover with a set and an item.

        Parameters:
        my_set (set): The original set.
        item (any): The item to be removed.
        """
        self.my_set = my_set
        self.item = item

    def remove_item(self):
        """
        Removes an item from the set using remove() or discard().

        Returns:
        set: The updated set after removing the item.
        """
        # Using discard to avoid error if item does not exist
        self.my_set.discard(self.item)
        return self.my_set


# Test the class
my_set = {1, 2, 3, 4, 5}
item_to_remove = 3

remover = SetItemRemover(my_set, item_to_remove)
result = remover.remove_item()
print(result)
```
