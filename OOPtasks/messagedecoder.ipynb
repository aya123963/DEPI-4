```python
class MessageDecoder:
    """
    MessageDecoder class handles the operation of decoding a message by:
    1. Reversing the first word.
    2. Keeping the second word as is (or modifying vowels if needed).

    Attributes:
    message (str): The input message to be decoded.
    """

    def __init__(self, message):
        """
        Initializes the MessageDecoder with the given message.

        Parameters:
        message (str): The input string/message entered by the user.
        """
        self.message = message

    def decode_message(self):
        """
        Decodes the message by reversing the first word and processing the second word.

        Returns:
        str: The decoded message with the first word reversed and the second word unchanged.
        """
        words = self.message.split()
        if not words:
            return ""

        # Reverse the first word
        first_word = words[0][::-1]

        # Process the second word (here we just keep it unchanged)
        second_word = words[1] if len(words) > 1 else ""

        # Combine words
        if second_word:
            return first_word + " " + second_word
        else:
            return first_word


# -------- Test the class --------
encoded_obj = MessageDecoder("emocleW EPGTQ")
decoded = encoded_obj.decode_message()
print(decoded)
```
