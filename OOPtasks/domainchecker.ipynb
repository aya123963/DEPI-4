```python
class DomainTypeChecker:
    """
    DomainTypeChecker class handles the operation of checking
    the type of domain ending of an email address.

    Attributes:
    email (str): The email address whose domain ending is checked.
    """

    def __init__(self, email):
        """
        Initializes the DomainTypeChecker with the given email.

        Parameters:
        email (str): The email address entered by the user.
        """
        self.email = email

    def check_domain_type(self):
        """
        Determines the domain type based on the email ending.

        Returns:
        str: The type of domain (Commercial, Educational, or Other).
        """
        if self.email.endswith(".com"):
            return "Commercial Domain"
        elif self.email.endswith(".edu"):
            return "Educational Domain"
        else:
            return "Other Domain"


# testing the class 
domain_checker = DomainTypeChecker("Amit_ml@gmail.edu")
result = domain_checker.check_domain_type()
print(result)
```
