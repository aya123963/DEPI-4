```python
class UsernameExtractor:
    """
    UsernameExtractor class handles the operation of extracting
    the username from an email address.

    Attributes:
    email (str): The email address from which the username is extracted.
    """

    def __init__(self, email):
        """
        Initializes the UsernameExtractor with the given email.

        Parameters:
        email (str): The email address entered by the user.
        """
        self.email = email

    def extract_username(self):
        """
        Extracts the username from the email address.

        Returns:
        str: The username part of the email.
        """
        return self.email.split("@")[0]


# testing the class with object from 
user = UsernameExtractor("Amit_ml@gmail.edu")
result = user.extract_username()
print(result)
```
