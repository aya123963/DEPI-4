```python
class SubsetChecker:
    """
    SubsetChecker class handles the operation of checking
    whether one set is a subset of another set.

    Attributes:
    set1 (set): The set to be checked.
    set2 (set): The set to compare against.
    """

    def __init__(self, set1, set2):
        """
        Initializes the SubsetChecker with two sets.

        Parameters:
        set1 (set): The potential subset.
        set2 (set): The main set.
        """
        self.set1 = set1
        self.set2 = set2

    def check_subset(self):
        """
        Checks if set1 is a subset of set2.

        Returns:
        bool: True if set1 is a subset of set2, otherwise False.
        """
        return self.set1.issubset(self.set2)



set1 = {1, 2}
set2 = {1, 2, 3, 4}

checker = SubsetChecker(set1, set2)
result = checker.check_subset()
print(result)
```
