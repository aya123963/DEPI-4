```python
class EvenNumberSummer:
    """
    EvenNumberSummer class handles the operation of summing
    all even numbers in a given list.

    Attributes:
    numbers (list of int): The list of numbers to process.
    """

    def __init__(self, numbers):
        """
        Initializes the EvenNumberSummer with a list of numbers.

        Parameters:
        numbers (list of int): The list of numbers.
        """
        self.numbers = numbers

    def sum_even_numbers(self):
        """
        Sums all even numbers in the list.

        Returns:
        int: The total sum of even numbers.
        """
        total = 0
        for num in self.numbers:
            if num % 2 == 0:
                total += num  # add if number is even
        return total


# Test the class
numbers = [1, 2, 3, 4, 5, 6]
summer = EvenNumberSummer(numbers)
result = summer.sum_even_numbers()
print(result)

```
