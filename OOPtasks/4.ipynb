```python
class SmallestFinder:
    """
    SmallestFinder class handles the operation of finding the smallest value in a list manually.

    Attributes:
    numbers (list of int/float): The list of numbers to search.
    """

    def __init__(self, numbers):
        """
        Initializes the SmallestFinder with a list of numbers.

        Parameters:
        numbers (list of int/float): The list of numbers.
        """
        self.numbers = numbers

    def find_smallest(self):
        """
        Finds the smallest value in the list without using min().

        Returns:
        int/float: The smallest number in the list.
        """
        smallest = self.numbers[0]
        for num in self.numbers:
            if num < smallest:
                smallest = num  # update smallest if a smaller number is found
        return smallest



numbers = [10, 3, 25, 1, 7]
finder = SmallestFinder(numbers)
result = finder.find_smallest()
print(result)
```
