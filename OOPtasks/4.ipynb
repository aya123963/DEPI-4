```python
class TupleIndexFinder:
    """
    TupleIndexFinder class handles the operation of finding
    the index of a specific item in a tuple.

    Attributes:
    my_tuple (tuple): The tuple to search in.
    item (any): The item whose index will be found.
    """

    def __init__(self, my_tuple, item):
        """
        Initializes the TupleIndexFinder with a tuple and an item.

        Parameters:
        my_tuple (tuple): The tuple to search.
        item (any): The item to find its index.
        """
        self.my_tuple = my_tuple
        self.item = item

    def find_index(self):
        """
        Finds the index of the given item in the tuple.

        Returns:
        int: The index of the item.
        """
        index = self.my_tuple.index(self.item)
        return index


# Test the class
my_tuple = ('apple', 'banana', 'orange', 'grape', 'banana')
item = 'orange'

finder = TupleIndexFinder(my_tuple, item)
result = finder.find_index()
print("index of 'orange' is:", result)

```
