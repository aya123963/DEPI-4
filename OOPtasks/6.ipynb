```python
class PalindromeChecker:
    """
    PalindromeChecker class handles the operation of checking
    whether a given string is a palindrome.

    Attributes:
    text (str): The string to be checked.
    """

    def __init__(self, text):
        """
        Initializes the PalindromeChecker with a string.

        Parameters:
        text (str): The string entered by the user.
        """
        self.text = text

    def is_palindrome(self):
        """
        Checks if the string is a palindrome.

        Returns:
        str: 'Palindrome' if the string is a palindrome, otherwise 'Not a Palindrome'.
        """
        if self.text == self.text[::-1]:
            return "Palindrome"
        else:
            return "Not a Palindrome"


# Test the class
text = input("Enter a string: ")
checker = PalindromeChecker(text)
result = checker.is_palindrome()
print(result)

```
