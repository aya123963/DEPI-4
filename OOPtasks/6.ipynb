```python
class WordLengthCounter:
    """
    WordLengthCounter class handles the operation of counting words
    that have a length greater than or equal to 2.

    Attributes:
    words (list of str): The list of words to be checked.
    """

    def __init__(self, words):
        """
        Initializes the WordLengthCounter with a list of words.

        Parameters:
        words (list of str): The list of words.
        """
        self.words = words

    def count_valid_words(self):
        """
        Counts the number of words with length >= 2.

        Returns:
        int: The count of words whose length is at least 2.
        """
        count = 0
        for word in self.words:
            if len(word) >= 2:
                count += 1  # increase counter when condition is met
        return count


# Testing  the class
words = ["hi", "a", "hello", "no", "on"]
counter = WordLengthCounter(words)
result = counter.count_valid_words()
print(result)
```
