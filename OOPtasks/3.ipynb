```python
class ListMultiplier:
    """
    ListMultiplier class handles the operation of multiplying all numbers in a list.

    Attributes:
    numbers (list of int/float): The list of numbers to be multiplied.
    """

    def __init__(self, numbers):
        """
        Initializes the ListMultiplier with a list of numbers.

        Parameters:
        numbers (list of int/float): The list of numbers.
        """
        self.numbers = numbers

    def multiply_all(self):
        """
        Multiplies all numbers in the list.

        Returns:
        int/float: The product of all numbers in the list.
        """
        result = 1
        for num in self.numbers:
            result *= num  # multiply each element
        return result


# Testingtheclass
numbers = [2, 3, 4]
multiplier = ListMultiplier(numbers)
result = multiplier.multiply_all()
print(result)
```
