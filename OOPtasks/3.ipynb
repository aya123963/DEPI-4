```python
class SetDifference:
    """
    SetDifference class handles the operation of finding the difference
    between two sets.

    Attributes:
    main_set (set): The original set.
    removed_set (set): The set of elements to be removed.
    """

    def __init__(self, main_set, removed_set):
        """
        Initializes the SetDifference with two sets.

        Parameters:
        main_set (set): The original set.
        removed_set (set): The set containing elements to remove.
        """
        self.main_set = main_set
        self.removed_set = removed_set

    def calculate_difference(self):
        """
        Calculates the difference between the two sets.

        Returns:
        set: A set containing elements that are in main_set but not in removed_set.
        """
        result = self.main_set - self.removed_set
        return result


# Test the class
my_set = {1, 2, 3, 4, 5}
removed = {1, 4}

diff = SetDifference(my_set, removed)
result = diff.calculate_difference()
print(result)

```
