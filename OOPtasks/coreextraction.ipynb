```python
import re

class CoreExtractor:
    """
    CoreExtractor class handles the operation of decoding a message
    and extracting only the core alphabetic parts from a string.

    Attributes:
    message (str): The input message from which core parts are extracted.
    """

    def __init__(self, message):
        """
        Initializes the CoreExtractor with the given message.

        Parameters:
        message (str): The input string/message entered by the user.
        """
        self.message = message

    def extract_core(self):
        """
        Extracts all alphabetic sequences from the message.

        Returns:
        list: A list of all alphabetic substrings found in the message.
        """
        return re.findall(r'[A-Za-z]+', self.message)


# -------- Test the class --------
core_obj = CoreExtractor("###!!@mocleW EPGTQ!!!6789")
result = core_obj.extract_core()
print(result)
```
