```python
class TupleModifier:
    """
    TupleModifier class handles the operation of adding an item to a tuple
    by converting it to a list and then back to a tuple.

    Attributes:
    my_tuple (tuple): The original tuple.
    item_to_add (any): The item to be added to the tuple.
    """

    def __init__(self, my_tuple, item_to_add):
        """
        Initializes the TupleModifier with a tuple and an item.

        Parameters:
        my_tuple (tuple): The original tuple.
        item_to_add (any): The item to be added.
        """
        self.my_tuple = my_tuple
        self.item_to_add = item_to_add

    def add_item(self):
        """
        Adds an item to the tuple by converting it to a list first.

        Returns:
        tuple: A new tuple containing the added item.
        """
        # Convert tuple to list
        temp_list = list(self.my_tuple)

        # Add the new item
        temp_list.append(self.item_to_add)

        # Convert back to tuple
        new_tuple = tuple(temp_list)

        return new_tuple


# testing the class to make sure it is working
my_tuple = (1, 2, 3)
item_to_add = 4

obj = TupleModifier(my_tuple, item_to_add)
result = obj.add_item()
print(result)
```
