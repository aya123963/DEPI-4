```python
class FibonacciSeries:
    """
    FibonacciSeries class handles generating Fibonacci numbers
    and creating a Fibonacci series up to a given number.

    Attributes:
    n (int): The number of terms in the Fibonacci series.
    """

    def __init__(self, n):
        """
        Initializes the FibonacciSeries with the number of terms.

        Parameters:
        n (int): Number of terms to generate.
        """
        self.n = n

    def fibonacci(self, n):
        """
        Calculates the Fibonacci number recursively.

        Parameters:
        n (int): Position in the Fibonacci sequence.

        Returns:
        int: The Fibonacci number at position n.
        """
        if n == 0:
            return 0
        if n == 1:
            return 1
        return self.fibonacci(n - 1) + self.fibonacci(n - 2)

    def generate_series(self):
        """
        Generates the Fibonacci series up to n terms.

        Returns:
        list: A list containing the Fibonacci series.
        """
        result = []
        for i in range(self.n):
            result.append(self.fibonacci(i))
        return result


# Test the class
fib = FibonacciSeries(10)
series = fib.generate_series()
print(series)

```
