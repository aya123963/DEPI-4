```python
class EmailValidator:
    """
    EmailValidator class handles the operation of validating an email address.

    Attributes:
    email (str): The email address to be validated.
    """
          # constructor --init--
    def __init__(self, email):
        """
        Initializes the EmailValidator with the given email.

        Parameters:
        email (str): The email address entered by the user.
        """
        self.email = email
          # Another method 
    def validate_email(self):
        """
        Checks whether the email is valid or invalid.

        Returns:
        str: 'Valid email' if the email format is correct, otherwise 'Invalid email'.
        """
        if self.email.count("@") == 1 and self.email.split("@")[1].count(".") >= 1:
            return "Valid email"
        else:
            return "Invalid email"


# Testing the code to make sure every thing is okay
email1 = EmailValidator("Amit_ml@gmail.edu")
result = email1.validate_email()
print(result)

```
