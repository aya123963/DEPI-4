```python
class FactorialCalculator:
    """
    FactorialCalculator class handles the operation of calculating
    the factorial of a given number using a for loop.

    Attributes:
    num (int): The number whose factorial is to be calculated.
    """

    def __init__(self, num):
        """
        Initializes the FactorialCalculator with a number.

        Parameters:
        num (int): The number to calculate factorial for.
        """
        self.num = num

    def calculate_factorial(self):
        """
        Calculates the factorial of the number.

        Returns:
        int: Factorial of the number.
        """
        factorial = 1
        for i in range(1, self.num + 1):
            factorial *= i  # multiply each number in the range
        return factorial


# Test the class
num = int(input("Enter a number: "))
calculator = FactorialCalculator(num)
result = calculator.calculate_factorial()
print(f"Factorial of {num} is {result}")

```
