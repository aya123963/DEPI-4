```python
class ListCloner:
    """
    ListCloner class handles the operation of creating a copy of a list
    without using direct assignment.
    
    Attributes:
    original (list): The original list to be copied.
    """

    def __init__(self, original):
        """
        Initializes the ListCloner with the original list.

        Parameters:
        original (list): The list to be copied.
        """
        self.original = original

    def clone_list(self):
        """
        Creates a copy of the list using slicing.

        Returns:
        list: A new list that is a copy of the original list.
        """
        cloned = self.original[:]  # create a copy using slicing
        return cloned


# Test the class
original = [1, 2, 3, 4]
cloner = ListCloner(original)
result = cloner.clone_list()
print(result)
```
