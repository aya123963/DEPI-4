```python
class LargestFinder:
    """
    LargestFinder class handles the operation of finding the largest value in a list manually.

    Attributes:
    numbers (list of int/float): The list of numbers to search.
    """

    def __init__(self, numbers):
        """
        Initializes the LargestFinder with a list of numbers.

        Parameters:
        numbers (list of int/float): The list of numbers.
        """
        self.numbers = numbers

    def find_largest(self):
        """
        Finds the largest value in the list without using max().

        Returns:
        int/float: The largest number in the list.
        """
        largest = self.numbers[0]
        for num in self.numbers:
            if num > largest:
                largest = num  # update largest if a bigger number is found
        return largest


# Testing the class by using its object
numbers = [10, 3, 25, 1, 7]
finder = LargestFinder(numbers)
result = finder.find_largest()
print(result)
```
