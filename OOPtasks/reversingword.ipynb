```python
class FirstWordDecoder:
    """
    FirstWordDecoder class handles the operation of decoding
    the first word of a message by reversing it.

    Attributes:
    message (str): The input message whose first word is decoded.
    """

    def __init__(self, message):
        """
        Initializes the FirstWordDecoder with the given message.

        Parameters:
        message (str): The input string/message entered by the user.
        """
        self.message = message

    def decode_first_word(self):
        """
        Reverses the first word of the message.

        Returns:
        str: The first word of the message reversed.
        """
        words = self.message.split()
        return words[0][::-1] if words else ""


# testing the class
encoded_obj = FirstWordDecoder("emocleW EPGTQ")
decoded = encoded_obj.decode_first_word()
print(decoded)
```
