```python
class DomainExtractor:
    """
    DomainExtractor class handles the operation of extracting
    the domain name from an email address.

    Attributes:
    email (str): The email address from which the domain is extracted.
    """

    def __init__(self, email):
        """
        Initializes the DomainExtractor with the given email.

        Parameters:
        email (str): The email address entered by the user.
        """
        self.email = email

    def extract_domain(self):
        """
        Extracts the domain name from the email address.

        Returns:
        str: The domain name without the top-level domain.
        """
        return self.email.split("@")[1].rsplit(".", 1)[0]


# testing the class by taking object
domain_obj = DomainExtractor("Amit_ml@gmail.edu")
result = domain_obj.extract_domain()
print(result)
```
