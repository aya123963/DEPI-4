```python
class ListSummer:
    """
    ListSummer class handles the operation of calculating the sum of numbers in a list.

    Attributes:
    numbers (list of int/float): The list of numbers to be summed.
    """

    def __init__(self, numbers):
        """
        Initializes the ListSummer with a list of numbers.

        Parameters:
        numbers (list of int/float): The list of numbers.
        """
        self.numbers = numbers

    def calculate_total(self):
        """
        Loops through the list and calculates the total sum.

        Returns:
        int/float: The total sum of all numbers in the list.
        """
        total = 0
        for num in self.numbers:
            total += num  # accumulate the sum
        return total


# Test the class
numbers = [5, 10, 15, 20]
summer = ListSummer(numbers)
result = summer.calculate_total()
print(result)
```
