```python
class SetRemover:
    """
    SetRemover class handles the operation of removing a specific item from a set.

    Attributes:
    my_set (set): The set from which an item will be removed.
    """

    def __init__(self, my_set):
        """
        Initializes the SetRemover with a set.

        Parameters:
        my_set (set): The original set.
        """
        self.my_set = my_set

    def remove_item(self, item):
        """
        Removes an item from the set using remove().

        Parameters:
        item (any): The item to be removed from the set.

        Returns:
        set: The updated set after removal.
        """
        self.my_set.remove(item)  # remove the specified item
        return self.my_set


# Test the class
my_set = {1, 2, 3, 4, 5}
remover = SetRemover(my_set)
result = remover.remove_item(3)
print(result)

```
