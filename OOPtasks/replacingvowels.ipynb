```python
class VowelShifter:
    """
    VowelShifter class handles the operation of decoding a message by:
    1. Reversing the first word.
    2. Replacing all vowels in the second word with 'A'.

    Attributes:
    message (str): The input message to be decoded.
    """

    def __init__(self, message):
        """
        Initializes the VowelShifter with the given message.

        Parameters:
        message (str): The input string/message entered by the user.
        """
        self.message = message

    def vowel_shift(self):
        """
        Decodes the message by reversing the first word and replacing vowels in the second word.

        Returns:
        str: The decoded message with the first word reversed and vowels in the second word replaced by 'A'.
        """
        words = self.message.split()
        if not words:
            return ""

        # Reverse the first word
        first_word = words[0][::-1]

        # Replace vowels in the second word
        vowels = "AEIOUaeiou"
        second_word = ""
        if len(words) > 1:
            for char in words[1]:
                if char in vowels:
                    second_word += "A"
                else:
                    second_word += char

        # Combine words
        if len(words) > 1:
            return first_word + " " + second_word
        else:
            return first_word


# -------- Test the class --------
encoded_obj = VowelShifter("emocleW EPGTQ")
decoded = encoded_obj.vowel_shift()
print(decoded)
```
