```python
#task
# Input validationing
x="Amit_ml@gmail.edu"
if x.count("@")==1 or x.split("@")[1].count(".") >= 1:
    print("Valid email")


else:
    print("Invalid email")
```

```python
# Extract Username
x="Amit_ml@gmail.edu"
email=x.split("@")[0]
print(email)
```

```python
#Extract Domain
x="Amit_ml@gmail.edu"
domain= x.split("@")[1].rsplit(".", 1)[0]
print(domain)
```

```python

# Checking for  domain ending
x="Amit_ml@gmail.edu"
if x.endswith(".com"):
    print("Commercial Domain")
elif x.endswith(".edu"):
    print("Educational Domain")
else:
    print("Other Domain")
```
